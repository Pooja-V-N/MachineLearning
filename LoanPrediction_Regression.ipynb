### Capping

```python
def out_capping(col):
    q3 = data[col].quantile(0.75)
    q1 = data[col].quantile(0.25)
    
    iqr = q3-q1
    
    upper = q3 + 1.5*iqr
    lower = q1 - 1.5*iqr
    
    data[col].clip(upper,lower,inplace = True)
```

```python
for i in num_col:
    out_capping(i)
```

```python
f,a = plt.subplots(2,4,figsize = (15,12),dpi = 100)

ind = 0
for i in range(2):
    for j in range(4):
        if ind < len(num_col):
            sns.boxplot(y= data[num_col[ind]],ax=a[i,j])
            a[i,j].set_title(num_col[ind])
            ind+=1
        else:
            a[i,j].axis('off') 

          
plt.tight_layout()
plt.show()
```
